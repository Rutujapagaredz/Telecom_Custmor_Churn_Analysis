```python
import matplotlib.pyplot as plt
import seaborn as sns

# ---------------------- Create Subplots ----------------------
fig, axes = plt.subplots(2, 2, figsize=(14, 10))  # 2 rows, 2 columns

# Churn rate percentage by pie chart (added as the first plot)
axes[0, 0].pie(df['Churn'].value_counts(), labels=df['Churn'].value_counts().index,
               autopct='%1.2f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
axes[0, 0].set_title("Churn Rate Percentage")

# Churn distribution (second plot)
ax = sns.countplot(x='Churn', data=df, ax=axes[0, 1])
ax.bar_label(ax.containers[0])
axes[0, 1].set_title('Churn Distribution')

# Churn by tenure group (third plot)
sns.countplot(x='tenure', hue='Churn', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Churn by Tenure Group')

# Customer Churn Count by Gender (fourth plot)
sns.countplot(x='gender', data=df, hue='Churn', palette='Set2', ax=axes[1, 1])
axes[1, 1].set_title('Customer Churn Count by Gender')
axes[1, 1].set_xlabel('Gender')
axes[1, 1].set_ylabel('Count')

# Adding labels to the bars in the Gender plot
for bars in axes[1, 1].containers:
    axes[1, 1].bar_label(bars)

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Create a count plot
ax = sns.countplot(x='SeniorCitizen', data=df, hue='Churn', palette='Set1')

# Set the title and labels
ax.set_title('Customer Churn Count by SeniorCitizen')
ax.set_xlabel('SeniorCitizen')
ax.set_ylabel('Count')

# Add labels to each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Show the plot
plt.show()
```

Customer Behavior: By seeing how the Churn counts compare across SeniorCitizen categories, you can gain insights into which group is more prone to churn.

Targeted Actions: If senior citizens have a higher churn rate, you might target this group with specific retention strategies, like offering them special deals or improved customer service.

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Create a count plot
ax = sns.countplot(x='Partner', data=df, hue='Churn', palette='Set1')

# Set the title and labels
ax.set_title('Customer Churn Count by Partner')
ax.set_xlabel('Partner')
ax.set_ylabel('Count')

# Add labels to each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Show the plot
plt.show()
```

You may observe that customers with partners tend to have a lower churn rate compared to those without a partner. If this is the case, it could suggest that having a partner might lead to more stable relationships with the company, and this group could require less retention effort.

If non-partnered customers are more likely to churn, you might focus on strategies to engage them more effectively, such as offering targeted promotions or additional benefits to encourage them to stay.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='tenure', data=df, palette='Set2')
plt.title('Tenure Distribution by Churn Status')
plt.xlabel('Churn')
plt.ylabel('Tenure (Months)')
plt.show()
```

```python
# Create tenure categories
df['TenureCategory'] = pd.cut(df['tenure'], bins=[0, 12, 24, 36, 48, 60, df['tenure'].max()], labels=['0-12', '12-24', '24-36', '36-48', '48-60', '60+'])

# Create a pivot table
churn_by_tenure = pd.crosstab(df['TenureCategory'], df['Churn'])

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(churn_by_tenure, annot=True, cmap='YlGnBu', fmt='d')
plt.title('Churn by Tenure Categories')
plt.xlabel('Churn')
plt.ylabel('Tenure Categories')
plt.show()
0
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# df = pd.read_csv('Churn_analysis.csv')  # Uncomment and update path if needed

# List of categorical service features
service_features = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up the matplotlib figure
plt.figure(figsize=(18, 12))

# Create 3x3 subplots
for i, feature in enumerate(service_features, 1):
    plt.subplot(3, 3, i)
    ax = sns.countplot(data=df, x=feature, hue='Churn', palette='Set1')
    plt.title(f'{feature} vs Churn', fontsize=14, fontweight='bold')
    plt.xlabel('')
    plt.ylabel('Count')
    plt.xticks(rotation=45)

    # Add value labels on bars
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', fontsize=9, color='black', xytext=(0, 5),
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Pnadas Task 2

# Group by service combinations and churn status
behavior_pattern = df.groupby([
    'PhoneService', 'InternetService', 'OnlineSecurity',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
])['Churn'].value_counts(normalize=True).unstack().fillna(0)

# View top patterns with high churn rate
behavior_pattern['Churn_Rate'] = behavior_pattern['Yes']
behavior_pattern_sorted = behavior_pattern.sort_values(by='Churn_Rate', ascending=False)
print(behavior_pattern_sorted.head(10))
```

```python
# Churn probability by InternetService type
churn_prob = df.groupby('InternetService')['Churn'].value_counts(normalize=True).unstack()
churn_prob['Churn_Probability'] = churn_prob['Yes']
print(churn_prob)
```

Fiber optic customers have the highest churn probability at 41.89%. This could suggest dissatisfaction with service, higher competition, or pricing issues.

DSL customers show a moderate churn rate of 18.96%, indicating relatively more stability compared to fiber optic users.

Customers with no internet service have the lowest churn rate (7.41%), which makes sense—they're likely less involved, possibly using only phone services.
